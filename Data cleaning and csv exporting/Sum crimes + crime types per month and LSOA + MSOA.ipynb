```python
import numpy as np
import matplotlib.pyplot as plt

import os
import datetime

import IPython
import IPython.display
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import tqdm as tq
import string
import statistics as stats
import re

from numpy import arange, mean, var, std, array, append, diff, cumsum

import sqlite3
```

```python
df = pd.read_csv('all_street.csv')
msoa = pd.read_csv('street_with_msoa.csv')
df = df.sort_values(by='Month')
df_copy = df.copy()
```

```python
df_copy = pd.get_dummies(df_copy, columns = ['Crime type'])
df_copy = df_copy.drop(['Unnamed: 0','Crime ID','LSOA name', 'Last outcome category', 'Falls within'], axis=1)
```

```python
df_types = df_copy.groupby(['Month', 'LSOA code']).sum()
df_sum = df.groupby(['Month', 'LSOA code']).count()
df_sum = df_sum[['Falls within']]
df_sum = df_sum.rename(columns= {'Crime type' : 'Sum crimes'})
df_all = pd.concat([df_types, df_sum], axis=1)
df_all.reset_index(inplace=True)
```

```python
months = df['Month'].unique()
lsoa = df['LSOA code'].unique()
```

```python
#make list of lsoa's in a certain month
lsoa_lst = []
for m in months:
    lst = []
    lst = df_all[df_all['Month'] == m]['LSOA code'].to_list()
    lsoa_lst.append(lst)
```

```python
# Implement zero values for months where no crimes happened in an lsoa
for ls, mo in zip(lsoa_lst, months):
    diff = set(ls).symmetric_difference(set(lsoa))
    df2 = pd. DataFrame({'LSOA code': list(diff)})
    df2['Month'] = mo
    df2['Crime type_Anti-social behaviour'] = 0
    df2['Crime type_Bicycle theft'] = 0
    df2['Crime type_Burglary'] = 0
    df2['Crime type_Criminal damage and arson'] = 0
    df2['Crime type_Drugs'] = 0
    df2['Crime type_Other crime'] = 0
    df2['Crime type_Other theft'] = 0
    df2['Crime type_Possession of weapons'] = 0
    df2['Crime type_Public order'] = 0
    df2['Crime type_Robbery'] = 0
    df2['Crime type_Shoplifting'] = 0
    df2['Crime type_Theft from the person'] = 0
    df2['Crime type_Vehicle crime'] = 0
    df2['Crime type_Violence and sexual offences'] = 0
    df2['Sum crimes'] = 0
    
    df2 = df2[['Month', 'LSOA code', 'Crime type_Anti-social behaviour',
 'Crime type_Bicycle theft',
 'Crime type_Burglary',
 'Crime type_Criminal damage and arson',
 'Crime type_Drugs',
 'Crime type_Other crime',
 'Crime type_Other theft',
 'Crime type_Possession of weapons',
 'Crime type_Public order',
 'Crime type_Robbery',
 'Crime type_Shoplifting',
 'Crime type_Theft from the person',
 'Crime type_Vehicle crime',
 'Crime type_Violence and sexual offences', 'Sum crimes']]
    df_all = pd.concat([df_all, df2], axis=0)
```

```python
df_sort = df_all.sort_values(by='Month')
df_sort.reset_index(inplace = True)
df_sort = df_sort.drop(['index'], axis=1)
```

```python
# Implement msoa
df_street_with_msoa = pd.merge(msoa[['LSOA code', 'MSOA code']], df_sort, how='right', on='LSOA code')
```

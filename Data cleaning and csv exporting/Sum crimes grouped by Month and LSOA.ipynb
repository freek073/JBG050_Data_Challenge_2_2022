```python
import numpy as np
import matplotlib.pyplot as plt

import os
import datetime

import IPython
import IPython.display
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import tqdm as tq
import string
import statistics as stats
import re

from numpy import arange, mean, var, std, array, append, diff, cumsum

import sqlite3
```

```python
df = pd.read_csv('all_street.csv')
```

```python
df['Datetime'] = pd.to_datetime(df['Month'], format='%Y.%m')
df['Month_int'] = pd.DatetimeIndex(df['Datetime']).month
df['Year_int'] = pd.DatetimeIndex(df['Datetime']).year
```

```python
df = df.sort_values(by='Month')
```

```python
df_grouped = df.groupby(['Month', 'LSOA code']).count()
```

```python
df_grouped
```

```python
df_grouped2 = df_grouped[['Crime type']]
```

```python
df_grouped2.reset_index(inplace=True)
```

```python
months = df['Month'].unique()
```

```python
lsoa = df['LSOA code'].unique()
```

```python
lsoa_lst = []
for m in months:
    lst = []
    lst = df_grouped2[df_grouped2['Month'] == m]['LSOA code'].to_list()
    lsoa_lst.append(lst)
```

```python
df_grouped_copy = df_grouped2.copy()
```

```python
for ls, mo in zip(lsoa_lst, months):
    diff = set(ls).symmetric_difference(set(lsoa))
    df2 = pd. DataFrame({'LSOA code': list(diff)})
    df2['Month'] = mo
    df2['Crime type'] = 0
    df2 = df2[['Month', 'LSOA code', 'Crime type']]
    df_grouped_copy = pd.concat([df_grouped_copy, df2], axis=0)
```

```python
df_sort = df_grouped_copy.sort_values(by='Month')
```

```python
df_sort = df_sort.reset_index()
```

```python
df_sort = df_sort[['Month', 'LSOA code', 'Crime type']]
```

```python
df_sort = df_sort.rename(columns= {'Crime type' : 'Sum crimes'})
```

```python
df_sort.to_csv('crimes_per_month_lsoa.csv')
```
